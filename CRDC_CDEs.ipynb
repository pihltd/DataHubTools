Get the collection of CRDC CDEs and their permissible values

```python
import requests
import pandas as pd
from IPython.display import display, Markdown, Latex
import json
```

```python
crdc_url="https://cadsrapi.cancer.gov/rad/NCIAPI/1.0/api/DataElements/getCRDCList"
```

```python
def getCRDCCDEs(url):
    headers = {"accept" : "application/json"}
    results = requests.get(url, headers=headers)
    return results.json()
```

```python
columns = ["CRDC Name", "Public ID", "Version", "Domain", "Registration", "Workflow","OwnedBy", "VDType"]
enum_columns = ["CRDC Name", "Public ID", "Version", "VM Long Name", "VM Public ID", "VM Concept Code"]
```

```python
cde_df = pd.DataFrame(columns = columns)
enum_df = pd.DataFrame(columns = enum_columns)
```

```python
crdc_json = getCRDCCDEs(crdc_url)
print(crdc_json)
```

```python
crdc_json = crdc_json['CRDCDataElements']
```

```python
for element in crdc_json:
    cde_df.loc[len(cde_df.index)] = [element['CRDC Name'],element['CDE Public ID'], element["Version"], element['CRD Domain'], element['Registration Status'], element['Workflow Status'],element['Owned By'], element['VD Type']]
    if element['VD Type'] == 'Enumerated':
        for pv in element['permissibleValues']:
            enum_df.loc[len(enum_df.index)] = [element['CRDC Name'],element['CDE Public ID'], element["Version"],pv['VM Long Name'], pv['VM Public ID'], pv['Concept Code']]
```
